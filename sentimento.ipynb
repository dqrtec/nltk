```python
import ast
import nltk

dados = open("dados.txt")
dados = ast.literal_eval( dados.read() )

stop_words = open("stopwords.txt")
stop_words = ast.literal_eval( stop_words.read() )

#stopwords do nltk
stop_words = nltk.corpus.stopwords.words("portuguese")
```

```python
def retira_steeming_stopword(frases, stop_words):
    stemmer = nltk.stem.RSLPStemmer()
    frases_sem_steen = []
    for (frase, emocao) in frases:
        frases_sem_steen.append( ([stemmer.stem(i) for i in frase.split() if i not in stop_words],emocao) )
    return frases_sem_steen
```

```python
def todos_radicais():
    frases = retira_steeming_stopword(dados, stop_words)
    todos = []
    for i,j in frases:
        for k in i:
            todos.append(k)
    return todos

def frequencia():
    a = todos_radicais()
    return nltk.FreqDist(a)

def palavras_unicas(distribuicao):
    return distribuicao.keys()

def mapa_ocorrencias(frases,unicas):
    mapa = []
    for frase, emocao in frases:
        mapa.append([])
        for radical in unicas:
            if(radical in frase):
                mapa[-1].append( 1 )
            else:
                mapa[-1].append( 0 )
    return mapa
```

```python
# pega as 4 palavras mais comuns
frequencia().most_common(4)
```

```python
todas_palavras_steeming = palavras_unicas( frequencia() )
```

```python
frases_steemin = retira_steeming_stopword(dados, stop_words)
```

```python
mapa_ocorrencias(frases_steemin, todas_palavras_steeming)
```
